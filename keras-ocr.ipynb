```python
import keras_ocr
import matplotlib.pyplot as plt
```

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

```python
# Read images from folder path to image object
# images = [
#     keras_ocr.tools.read(img) for img in [r'D:\Surya\College\Major Project\OCR_model\dataset\test_v2\test\A_test_a.jpg',
#                                           r'D:\Surya\College\Major Project\OCR_model\dataset\test_v2\test\A_test_image_1.jpg',
#                                           r'D:\Surya\College\Major Project\OCR_model\dataset\test_v2\test\A_test_image.jpg']
# ]

images = [
    keras_ocr.tools.read(img) for img in [r'me0h2wpdvgk91.jpg',
                                          r'D:\Surya\College\Major Project\OCR_model\dataset\test_v2\test\A_test_image.jpg']
]
```

```python
prediction_groups = pipeline.recognize(images)
```

```python
fig, axs = plt.subplots(nrows=len(images), figsize=(10, 20))
for ax, image, predictions in zip(axs, images, prediction_groups):
    keras_ocr.tools.drawAnnotations(image=image, 
                                    predictions=predictions, 
                                    ax=ax)
```
